```python
from PyPDF2 import PdfReader

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
            cleaned_text = text.replace('□\t', '\n')
            
    return cleaned_text

def save_text_to_file(text, output_file):
    with open(output_file, "w") as file:
        file.write(text)

# Get the user's home directory
home_dir = os.path.expanduser("~")

# Define PDF paths and corresponding output file paths in a list of tuples
pdf_paths = [
    ('../Datasets/Sample_PDFs/2024-l1-topics-combined-2.pdf', '../Datasets/PyPDF/PyPDF_RR_2024_l1_combined.txt'),
    ('../Datasets/Sample_PDFs/2024-l2-topics-combined-2.pdf', '../Datasets/PyPDF/PyPDF_RR_2024_l2_combined.txt'),
    ('../Datasets/Sample_PDFs/2024-l3-topics-combined-2.pdf', '../Datasets/PyPDF/PyPDF_RR_2024_l3_combined.txt')
]


# Iterate over the list of PDF paths and output file paths
for i, (pdf_path, output_file) in enumerate(pdf_paths, start=1):
    print(f"Extracted text from PDF {i} saved to {output_file}")
    pdf_text = extract_text_from_pdf(pdf_path)
    save_text_to_file(pdf_text, output_file)
    
```

```python
import requests
import json
import os
from xml.etree import ElementTree as ET

# Get the user's home directory
home_dir = os.path.expanduser("~")

# Define PDF paths and corresponding output file paths in a list of tuples
file_paths = [
    ('../Datasets/Sample_PDFs/2024-l1-topics-combined-2.pdf',
     '../Datasets/Grobid/Grobid_RR_2024_l1_combined.xml',
     '../Datasets/Grobid/Grobid_RR_2024_l1_combined.txt'),

    ('../Datasets/Sample_PDFs/2024-l2-topics-combined-2.pdf',
     '../Datasets/Grobid/Grobid_RR_2024_l2_combined.xml',
     '../Datasets/Grobid/Grobid_RR_2024_l2_combined.txt'),

    ('../Datasets/Sample_PDFs/2024-l3-topics-combined-2.pdf',  
     '../Datasets/Grobid/Grobid_RR_2024_l3_combined.xml',
     '../Datasets/Grobid/Grobid_RR_2024_l3_combined.txt'),
]
```

```python
# GROBID endpoint URL
grobid_url = "http://localhost:8070/api/processFulltextDocument"


for pdf_path, xml_path, txt_path in file_paths:
    # Read the PDF file as binary
    with open(pdf_path, "rb") as file:
        pdf_content = file.read()

    # Send a POST request to GROBID endpoint to process the PDF
    response = requests.post(grobid_url, files={'input': pdf_content,
                                                 'consolidateHeader': 1,
                                                 'consolidateCitations': 1,
                                                 'consolidateFunders': 1,
                                                 'includeRawCitations': 1,
                                                 'includeRawAffiliations': 1,
                                                 'segmentSentences': 1,
                                                 'start': 1})

    # Check if the request was successful
    if response.status_code == 200:
        # Extracted content from PDF
        extracted_content = response.text
        print("Content extracted from", pdf_path)
        
        # Write the extracted content to an XML file
        with open(xml_path, "w") as xml_file:
            xml_file.write(extracted_content)
        print("Extracted content stored in:", xml_path)
        
        # Parse XML file
        tree = ET.fromstring(extracted_content)

        # Extract text content from XML elements
        text_content = [element.text.strip() for element in tree.iter() if element.text]

        # Join the text content into a single string
        full_text = '\n'.join(text_content)
        
        full_text = full_text.replace(' □ ', '\n')
        
        # Write the text content to a text file
        with open(txt_path, "w") as text_file:
            text_file.write(full_text)

        print("Text content extracted and stored in:", txt_path)
    else:
        print("Failed to extract content from", pdf_path)
```

# Structured metadata using Grobid for XML files to a CSV file

```python
import os
import re
import csv
import requests
import xml.etree.ElementTree as ET
from io import BytesIO
import pandas as pd 

# Define a function to extract metadata from an XML file
def extract_metadata(xml_path):
    metadata = {}

    # Load the TEI XML file
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Extract TEI Header metadata
    tei_header = root.find(".//{http://www.tei-c.org/ns/1.0}teiHeader")
    if tei_header is not None:
        metadata['language'] = tei_header.get('{http://www.w3.org/XML/1998/namespace}lang')

    # Extract version and encoding
    with open(xml_path, 'r', encoding='utf-8') as f:
        first_line = f.readline()
        match = re.match(r'^<\?xml\s+version\s*=\s*(["\'])(.*?)\1\s+encoding\s*=\s*(["\'])(.*?)\3', first_line)
        if match:
            metadata['version'] = match.group(2)
            metadata['encoding'] = match.group(4)

    # Get file size in bytes
    file_stats = os.stat(xml_path)
    f_size = file_stats.st_size
    metadata['file_size'] = f_size

    return metadata

metadata_dict = {'File_Name':[], 'language':[],'version':[],'encoding':[],'file_size':[], 'S3_link': []}

pdf_links = [
    "https://cfadatabase.s3.amazonaws.com/PyPDF_Grobid_RR_2024_l1_combined.txt",
    "https://cfadatabase.s3.amazonaws.com/PyPDF_Grobid_RR_2024_l2_combined.txt",
    "https://cfadatabase.s3.amazonaws.com/PyPDF_Grobid_RR_2024_l3_combined.txt"
]

# Iterate through each XML file path in Metadata_xml
for _,xml_path,_ in file_paths:
    metadata_dict['File_Name'].append(xml_path)
    # Extract metadata for the current XML file
    metadata = extract_metadata(xml_path)
    for key in metadata.keys():
        metadata_dict[key].append(metadata[key])
    # Add S3 link
    metadata_dict['S3_link'].append(pdf_links.pop(0))  # Pop the first link from the list
    
    # Print extracted metadata
    print("Metadata for", xml_path, ":", metadata)
    
df=pd.DataFrame(metadata_dict)
df.to_csv('../Datasets/Grobid/metadata.csv')
```
